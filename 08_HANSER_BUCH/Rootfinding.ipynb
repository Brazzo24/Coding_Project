```python
from scipy.optimize import newton
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x):
    return (x+1)*(x-3)*(x-5)
x = np.linspace(-2,6,80)
y = f(x)
x1 = newton(f,-2)
x2 = newton(f,2)
x3 = newton(f,6)
print ("Die Nullstellen sind: ",x1,x2,x3)
nullstellen = np.array([x1,x2,x3])
plt.scatter(nullstellen,[0,0,0],s=100)
plt.plot(x,y,'r-', linewidth=2)
plt.xlabel("x-Werte")
plt.ylabel("y-Werte")
plt.title("Nullstellen")
plt.axhline(y=0.5, color='black', linestyle='-')
plt.show()
plt.grid(True)
plt.show()
```
