```python
from scipy.interpolate import interp1d
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(1,6,6)
y = np.array([1,3,1,5,4,4])
xn = np.linspace(1,6,60)
f = interp1d(x,y,kind='linear')
g = interp1d(x,y,kind='cubic')
h = interp1d(x,y,kind=5)
plt.plot(x,y,'ro',linewidth=3.0)
plt.plot(xn,f(xn),'b-',linewidth=3)
plt.plot(xn,g(xn),'g--',linewidth=3)
plt.plot(xn,h(xn),'co-',linewidth=3)
plt.xlabel("x-Werte")
plt.ylabel("y-Werte")
plt.title("Interpolation")
plt.legend(["Vorgabe","Linear","Kubisch","Polynom Grad 5"], loc="best")
plt.show()

```
