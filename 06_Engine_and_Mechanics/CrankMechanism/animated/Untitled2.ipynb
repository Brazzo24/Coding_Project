```python
import matplotlib.pyplot as plt
from scipy.fftpack import fft
import numpy as np
from math import pi
```

```python
#generate sine wave
Fs = 300
t = np.arange(0,0.5,1/Fs)
f = 20;

x = np.sin(2*pi*f*t)+1.0*np.sin(2*pi*2*f*t) + 1.0*np.sin(2*pi*0.5*f*t)+ 1.4*np.sin(2*pi*1.5*f*t)+ 0.2*np.sin(2*pi*2.5*f*t)

plt.subplot(2,1,1)
plt.plot(t,x); plt.title('Sinusoidal Signal');
plt.xlabel('Time(s)'); plt.ylabel('Amplitude')
```
