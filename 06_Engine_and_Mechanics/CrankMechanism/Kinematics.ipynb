```python
#targets for this calculation
#receive piston position
#get coordiantes of BE and SE for requested crankangle

import numpy as np
from matplotlib import pyplot as plt
```

```python
#ipnut parameters
r = 0.02425 #crank radius in m
l = 0.0902 #conrod length in m
phi = 20 #kurbelwinkel

```

```python
def function_Piston_Pos(r,l,phi):
    
    x = l+r-np.sqrt(l**2-r**2*np.sin(phi*np.pi/180)**2)-r*np.cos(phi*np.pi/180)
    
    return x
```

```python

x_phi = []

for phi in range(0,360):
    x = function_Piston_Pos(r,l,phi)
    x_phi.append(x)
```

```python
print(x_phi[180])
```

```python
plt.plot(x_phi)
plt.grid()
plt.title('Piston Postion in [m]')
```
