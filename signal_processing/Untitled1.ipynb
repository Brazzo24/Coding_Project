```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#sine wave with given sample rate and frequency
S_rate = 44100 #Hz
T = 1/S_rate # time period of one sample
t = 0.1 #second tie duration of the signal
N = S_rate * t #number of samples needed for duration desire with t
```

```python
freq = 100 #Hz
omega = 2*np.pi*freq
```

```python
t_seq = np.arange(N)*T #sequence of point for N
```

```python
y = np.sin(omega*t_seq)
```

```python
plt.plot(t_seq,y)
plt.show()
```
